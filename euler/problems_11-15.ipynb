- '우수평' 방향으로 연속된 수 네 개의 곱 구하기
- '하수직' 방향으로 연속된 수 네 개의 곱 구하기
- '대각선' 방향으로 연속된 수 네 개의 곱 구하기
    - 대각선 조건에 트릭이 있는데 바로 '우하향' 대각선 뿐만 아니라 '좌하향' 대각선도 고려해야 한다는 것
    - '좌수평' 방향과 '상수직' 방향은 고려하지 않아도 되는 이유는 격자의 모든 값을 [0][0]부터 순서대로 순회하므로 결국 특정 [i][j]에서 계산한 '우수평' 방향의 값은 [i][j:j+4]에서 구하는 '좌수평' 방향의 값과 동일하고 '하수직' 방향의 값은 [i:i-4][j]에서 구하는 '상수직' 방향의 값과 동일하기 때문이다.
    

```python
from typing import List
from functools import reduce

def multiply(arr: List[int]) -> int:
    return reduce(lambda x, y: x*y, arr)

board = input().split()
board = [list(map(int, board[i*20:i*20+20])) for i in range(20)]

ans = 0

for i in range(20):
    for j in range(20):
        
        # 수평 방향으로 연속된 수 네 개의 곱
        if  j+3 < 20:
            arr = board[i][j:j+4]
            tmp_ans = multiply(arr)
            if tmp_ans > ans:
                ans = tmp_ans
        
        # 수직 방향 연속된 수 네 개의 곱
        if i+3 < 20:
            arr = [board[i+k][j] for k in range(4)]
            tmp_ans = multiply(arr)
            if tmp_ans > ans:
                ans = tmp_ans
        
        # 우하향 대각선 방향으로 연속된 수 네 개의 곱
        if i+3 < 20 and j+3 < 20:
            arr = [board[i+k][j+k] for k in range(4)]
            tmp_ans = multiply(arr)
            if tmp_ans > ans:
                ans = tmp_ans

        # 좌하향 대각선 방향으로 연속된 수 네 개의 곱
        if i+3 < 20 and 0 <= j-3 < 20:
            arr = [board[i+k][j-k] for k in range(4)]
            tmp_ans = multiply(arr)
            if tmp_ans > ans:
                ans = tmp_ans

print('정답:', ans)
```
